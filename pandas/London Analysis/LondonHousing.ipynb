# Q1

# Q2 

Add a new column 'Year' in the dataframe which contains years only

```python
london['Year'] = london.date.dt.year
```

```python
london
```

# Q3

Add a month as a new column, between date and area column

```python
london.insert(1,'Month',london.date.dt.month )
```

# Q4

Remove the column Year and month from the Dataframe

```python
london.drop(['Year', 'Month'], axis = 1, inplace = True)
```

# Q6

What is the maximum and minimum 'avarage price' per year in england


```python
london
```

```python
london['Year'] = london.date.dt.year
```

# Q6 

what is the Average max and min price per year in england

```python
df1.groupby('Year').average_price.max()
```

# Q7

what is the max and min no of crimes recorded area ?

```python
london.groupby('area').no_of_crimes.max()
```

```python
london.groupby('area').no_of_crimes.min().sort_values(ascending = True)
```
